```python
#데이터프레임을 생성한 후 데이터프레임의 정보를 출력하시오.
#대표국적별 영화 편수를 집계하시오.
#대표국적이 ‘영국’인 데이터를 출력하시오.
#대표국적별 최상위 순위를 출력하시오.

df = pd.read_excel('BoxOffice.xlsx')
# 데이터 출력
print(df)
df.info()
# 대표국적별 영화 편수 집계
nation_counts = df['대표국적'].value_counts()
for nation, count in nation_counts.items():
    print(f"{nation}: {count}")
# 대표국적이 '영국'인 데이터를 출력
uk_rows = df[df.iloc[:, 7] == '영국']
print(uk_rows)
# 대표국적별 최상위 순서 출력
first_occurrence = {}
for index, row in df.iterrows():
    nation = row.iloc[7]
    if nation not in first_occurrence:
        first_occurrence[nation] = index + 1
for nation, row_number in first_occurrence.items():
    print(f"{nation}: {row_number}")
```
